```python
%load_ext autoreload
%autoreload 2
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.linear_model import LogisticRegression
```

```python
db = ["HAM"]
models = ["RES", "DENSE", "MOBILE"]
root = f"../../dataset/CAI 2025/{db[0]}/"

filepaths = {}
for i in range(len(models)):
    filepaths[models[i]] = f"{root}eval/df_valid_{models[i]}.csv"

df = pd.DataFrame()
for k, v in filepaths.items():
    tmp = pd.read_csv(v)
    tmp["model"] = k
    tmp["db"] = db[0]
    df = pd.concat([df, tmp])
```

```python
df["accuracy"] = 0
df.loc[df["label"] == df["pred"], "accuracy"] = 1
```

```python
X = df[['WD']]
y = df['accuracy']
model = LogisticRegression()
model.fit(X, y)

X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_prob = model.predict_proba(X_range)[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='WD', y='accuracy', alpha=0.5)
plt.plot(X_range, y_prob, color='red', linewidth=2)
plt.title('Logistic Regression Fit')
plt.xlabel('WD')
plt.ylabel('Accuracy')
plt.grid()
plt.xlim(X['WD'].min(), X['WD'].max())
plt.ylim(-0.1, 1.1) 
plt.axhline(0.5, color='gray', linestyle='--') 
plt.show()
```

```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
from skincolors import IndividualTypologyAngle
```

```python
db = "HAM"
root = f"../../dataset/CAI 2025/{db}/"
df_valid = pd.read_csv(f"{root}eval/df_valid_RES.csv")
```

```python
df_valid["accuracy"] = 0
df_valid.loc[df_valid["label"] == df_valid["pred"], "accuracy"] = 1
```

```python
image_vectors = []

for i, row in df_valid.iterrows():
    img = cv2.imread(row["filepath"])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    image_vectors.append(img.flatten())

image_vectors = np.array(image_vectors)
```

# 正解・不正解

```python
pca = PCA(n_components=2)
results = pca.fit_transform(image_vectors)
colors = ["blue" if acc == 1 else "red" for acc in df_valid["accuracy"]]
plt.figure(figsize=(10, 7))
plt.scatter(results[:, 0], results[:, 1], c=colors, alpha=0.4)
plt.title("PCA Result")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid()
plt.show()
```

```python
pca = PCA(n_components=3)
results = pca.fit_transform(image_vectors)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(results[:, 0], results[:, 1], results[:, 2], c=colors, alpha=0.4)
ax.set_title("PCA Result in 3D")
ax.set_xlabel("Component 1")
ax.set_ylabel("Component 2")
ax.set_zlabel("Component 3")
plt.grid()
plt.show()
```

```python
tsne = TSNE(n_components=2, random_state=42)
results = tsne.fit_transform(image_vectors)
plt.figure(figsize=(10, 7))
plt.scatter(results[:, 0], results[:, 1], c=colors, alpha=0.4)
plt.title("t-SNE Result")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid()
plt.show()
```

```python
tsne = TSNE(n_components=3, random_state=42)
results = tsne.fit_transform(image_vectors)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(results[:, 0], results[:, 1], results[:, 2], c=colors, alpha=0.4)
ax.set_title("TSNE Result in 3D")
ax.set_xlabel("Component 1")
ax.set_ylabel("Component 2")
ax.set_zlabel("Component 3")
plt.grid()
plt.show()
```

```python
umap_model = umap.UMAP(n_components=2, random_state=42)
results = umap_model.fit_transform(image_vectors)
plt.figure(figsize=(10, 7))
plt.scatter(results[:, 0], results[:, 1], c=colors, alpha=0.4)
plt.title("UMAP Result")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid()
plt.show()
```

```python
umap_model = umap.UMAP(n_components=3, random_state=42)
results = umap_model.fit_transform(image_vectors)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(results[:, 0], results[:, 1], results[:, 2], c=colors, alpha=0.4)
ax.set_title("UMAP Result in 3D")
ax.set_xlabel("Component 1")
ax.set_ylabel("Component 2")
ax.set_zlabel("Component 3")
plt.grid()
plt.show()
```
